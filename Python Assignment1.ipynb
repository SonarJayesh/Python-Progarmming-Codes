```python
str4 = input("Enter the value: ")
vowels = consonants = digits = spaces = 0
for ch in str4:
    if ch.lower() in 'aeiou':
        vowels += 1
    elif ch.isalpha():
        consonants += 1
    elif ch.isdigit():
        digits += 1
    elif ch.isspace():
        spaces += 1
print("vowels: ",vowels)
print("consonants: ",consonants)
print("digits: ",digits)
print("spaces: ",spaces)
```

```python
# 10.Python Program to find simple interest and compound interest

principal = float(input("Principal: "))
rate = float(input("Rate: "))
time = float(input("Time: "))

simInterest = (principal * rate * time) / 100
compInterest = principal * ((1 + rate/100)**time - 1)

print("Simple Interest =", simInterest)
print("Compound Interest =", compInterest)
```

```python
# 11.Python program to check whether a number is Prime or not

num1 = int(input("Enter The number: "))
if num1 > 1:
    for i in range(2, num1):
        if num1 % i == 0:
            print("Not Prime")
            break
    else:
        print(num1,"is Prime")
else:
    print(num1, "is Not Prime")

```

```python
# 12.Python program to Count Even and Odd numbers in a string

str6 = input("Enter a string: ")
even = odd = 0
for ch in str6:
    if ch.isdigit():
        if int(ch) % 2 == 0:
            even += 1
        else:
            odd += 1
print("Even:", even, "Odd:", odd)

```

```python
# 18.Write a Python function to reverses a string if its length is a multiple of 5

def reverse_mult_5(str12):
    if len(str12) % 5 == 0:
        return str12[::-1] 
    else:
        return str12 

str12 = input("Enter a string: ")

print("Result:",reverse_mult_5(str12))
```
